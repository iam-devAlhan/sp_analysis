## Ghana Cloud Cover Data Assessment

In this assessment, we have extracted some data from the api done in stac_analysis.py
For the api, we have got the data required for analysis and exported in CSV for further

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../ghana_cloud_cover_2025.csv")

df['datetime'] = pd.to_datetime(df['datetime'], utc=True, errors="coerce")

df['month'] = df['datetime'].dt.month

monthly_cloud = df.groupby('month')["cloud_cover"].mean()

df.head(20)
```

```python

plt.figure(figsize=(8,5))
plt.plot(monthly_cloud.index, monthly_cloud.values, marker="o", color="teal")
plt.xticks(monthly_cloud.index)

plt.xlabel("Months from June to November (2025)")
plt.ylabel("Average Cloud Cover (%)")

plt.title("Monthly Average Cloud Cover over Ghana")
plt.grid(True)
plt.savefig("Average_cloud_cover_ghana_2025_line.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(monthly_cloud.index, monthly_cloud.values, color="teal")
plt.xticks(monthly_cloud.index)

plt.xlabel("Months from June to November (2025)")
plt.ylabel("Average Cloud Cover (%)")

plt.title("Monthly Average Cloud Cover over Ghana")
plt.savefig("Average_cloud_cover_ghana_2025_bar.png")
plt.show()
```

## In this Analysis, we covered average cloud cover in the data given as fetched from the API given.

> # Line Graph Analysis
The Line Graph states that the 10th Month or 11th Month that is October or November is best suitable for taking cloud satellite images over Ghana due to less cloud cover percentage
